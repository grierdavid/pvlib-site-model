```python
import numpy as np
num_variations = 10
solar_curves = {}
for i in range(num_variations):
    # Copy base weather data to perturb
    weather_perturbed = clearsky.copy()

    # Apply variability (5% standard deviation noise)
    for col in ['ghi', 'dhi', 'dni']:
        noise = np.random.normal(1.0, 0.05, size=len(clearsky))
        weather_perturbed[col] *= noise
        weather_perturbed[col] = weather_perturbed[col].clip(lower=0)

    # Run your existing PlantModel on perturbed weather
    mc_results = mc.run_model(weather_perturbed)
    
    # Extract AC output and convert W to MW
    ac_output_mw = mc_results.results.ac / 1e6

    solar_curves[f'solar_curve_{i+1}'] = ac_output_mw.values

# Create dataframe for all curves
solar_df = pd.DataFrame(solar_curves, mc_results.results.times)

# Save result to CSV
output_path = 'synthetic_solar_curves.csv'
solar_df.to_csv(output_path)

print(f"\nSuccessfully generated {num_variations} solar curves and saved to {output_path}")

# Optional quick plot
solar_df.iloc[:24*7].plot(figsize=(12,6), title="First Week of Synthetic Solar Curves")
```
